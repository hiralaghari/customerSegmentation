```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from kmodes.kprototypes import KPrototypes
from sklearn import metrics
from scipy.spatial.distance import cdist
```

```python
df = pd.read_csv('customer_data.csv')
```

```python
df.head()
```

#### EDA and Statistical Analysis

```python
df.info()
```

```python
df.describe()
```

```python
num_cols = ['age', 'income', 'purchase_amount']
cat_cols = [i for i in df.columns if i not in num_cols]
```

```python
fig=plt.figure(figsize=(16,4))
for i, col in enumerate(num_cols):
    ax=fig.add_subplot(1,3,i+1)
    sns.histplot(df[col], stat='density', kde=True, kde_kws={"cut": 3})
    ax.set_title(col+" Distribution")

fig.tight_layout()  
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(20, 10))  # Adjust figure size as needed

axes = axes.flatten()
for i, col in enumerate(cat_cols[1:]):
    sns.countplot(x=col, data=df, ax=axes[i])
    axes[i].set_title(f'Count Plot of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x')

fig.tight_layout()  
plt.show()
```

```python
sns.pairplot(df, vars=num_cols)
```

```python
matrix = df[num_cols].corr()
plt.imshow(matrix)
plt.colorbar()
variables = []
for i in matrix.columns:
    variables.append(i)
plt.xticks(range(len(matrix)), variables)
plt.yticks(range(len(matrix)), variables)
# Display the plot
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(20, 10))  # Adjust figure size as needed

axes = axes.flatten()
for i, item in enumerate(cat_cols[1:-2]):     
    sns.boxplot(x = df['age'], y=df[item], ax=axes[i])
    axes[i].set_title(f'Box Plot of {item}')

fig.tight_layout()  
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(20, 10))  # Adjust figure size as needed

axes = axes.flatten()
for i, item in enumerate(cat_cols[1:-2]):     
    sns.boxplot(x = df['purchase_amount'], y=df[item], ax=axes[i])
    axes[i].set_title(f'Box Plot of {item}')

fig.tight_layout()  
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(20, 10))  # Adjust figure size as needed

axes = axes.flatten()
for i, item in enumerate(cat_cols[1:-2]):     
    sns.boxplot(x = df['income'], y=df[item], ax=axes[i])
    axes[i].set_title(f'Box Plot of {item}')

fig.tight_layout()  
plt.show()
```
